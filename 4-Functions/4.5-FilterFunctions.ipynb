The filter() in python:

The filter() constructs an iterator from elements of an iterable for which a function returns true.It is used to filter out items from a list(or any other iterable) based on condition.

```python
#Filter function
def is_even(number):
    if number%2==0:
        return True

```

```python
numbers=[1,2,3,4,5,6,8,10,12,14,16]
list(filter(is_even,numbers))
```

```python
#Filter with lambda function with multiple conditions
lst=[1,2,3,4,5,6,19,28,34,37,31,44,43,64,65]

is_even_numbers=list(filter(lambda x:x%2==0 and x>10,lst))
print(is_even_numbers)
```

```python
#Filter to check if the age is greaterthan 25 in the dictionaries with lambda function
people=[{'name':"Tilak","age":17},
        {'name':"Syamala","age":29},
        {'name':"Krishna Kumari","age":47},
        {'name':"Srinivasa Rao","age":57}
        ]
is_age_greaterthan_25=list(filter(lambda person:person['age']>25,people))
print(is_age_greaterthan_25)
```

```python
def age_greaterthan_25(person):
    return person['age']>25

print(list(filter(age_greaterthan_25,people)))
```

Conclusion:


The filter function is powerful tool for creating iterator that filter items out of an iterable based on a function.It is commonly used for data cleaning,filtering objects, and removing unwanted elements from lists by mastering filter(),you can write more concise and efficient code for processing and manupulating collections in python.
