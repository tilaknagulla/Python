Conditional Statements(if else if else):



```python
#if statement 
age = 20
if age>=18:
    print("You are eligible to vote.")
    
```

```python
#else statement
#The else statement is used to execute a block of code when the condition in the if statement is false. It provides an alternative path of execution when the condition is not met.
age = 16
if age>=18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

```python
# else if statement
#The elif statement is used to check multiple conditions in a sequence. It allows you to specify additional conditions to be checked if the previous conditions are false. The elif statement stands for "else if" and is used to create a chain of conditions.
age = 17
if age<=13:
    print("You are a child.")
elif age<=18:
    print("You are a teenager.")
else:
    print("You are an adult.")
```

```python
#Nested conditional statements
#Nested conditional statements are if statements that are contained within another if statement. They allow you to  create more complex decision-making structures by checking multiple conditions at different levels of nesting. The inner if statements are evaluated only if the outer if statement's condition is true.
num = int(input("Enter a number:"))
if num>0:
    if num%2==0:
        print(num,"Is is even number")
    else:        
        print(num,"Is odd number")
else:   
    print(num,"Is is negative number")
```

```python
# Practical Example: 
##Determine if the year is a leap year or not. A leap year is defined as follows:
##- It is divisible by 4;   
##- However, if it is divisible by 100, it must also be divisible by 400 to be a leap year.
year = int(input("Enter a year:"))
if year%4==0:
    if year%100==0:
        if year%400==0:
         print(year,"is a leap year.")
        else:
           print(year,"Is not a leap year")
    else:
        print(year,"is a leap year.")
else:
    print(year,"Is not a leap year")
```

```python
#Simple calculator using nested if,elif and else
num1=int(input("Enter first number:"))
num2=int(input("Enter second number:"))
operator=input("Enter operator (+,-,*,/):")
if operator=="+":
    print("Addition opertaion",num1+num2)
elif operator=="-":
    print("Substration operator:",num1-num2)
elif operator=="*":
    print("Multiplication operator:",num1*num2)
elif operator=="/":
    print("Division operator:",num1/num2)
else:
    print("Invalid operator")
```
