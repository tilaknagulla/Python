### List Comprehension

Basic Syntex [expresiion for item in iterable]
with conditional logic [expression for item in iterable if condition]

```python
### Basic List Comprehension
square=[num**2 for num in range(10)]
print(square)  
```

```python
### List Comprehension with when condition
lst=[]
for x in range(10):
    if x%2==0:
     lst.append(x)

print(lst)
```

```python
even_numbers=[num for num in range(10) if num%2==0]
print(even_numbers)
```

```python
###Nested list comprehension 
lst1=[1,2,3,4,5]
lst2=['a','b','c','d','e']
pair=[[i,j] for i in lst1 for j in lst2]
print(pair)
```

```python
###List comprehension with word in words
words=["apple","orange","banana","pomogranite"]

word_length=[len(word) for word in words]
print(word_length)
```
