Operators:
=========

```python
## Arthmetic Operators
a=10
b=5
add_result=a+b # Addition
sub_result=a-b # Subtraction
mul_result=a*b # Multiplication 
div_result=a/b # Division
mod_result=a%b # Modulus   
floor_div_result=a//b # Floor Division
exp_result=a**b # Exponentiation
print("Addition:",add_result)       
print("Subtraction:",sub_result)
print("Multiplication:",mul_result)
print("Division:",div_result)
print("Modulus:",mod_result)
print("Floor Division:",floor_div_result)
print("Exponentiation:",exp_result)
```

```python
##Comparison Operators
a=10
b=10
a==b # Equal to
```

```python
str1="Tilak"
str2="tilak"
str1==str2 # Equal to
```

```python
# not equal to operators !=
str1="Tilak"
str2="tilak"
str1!=str2 # Not equal to
```

```python
# Greater than operator 
num1=45
num2=30

num1>num2 # Greater than
```

```python
#lessthan operator
print(num1<num2) # Less than
```

```python
## Gereater than or equal to operator >=
print(num1>=num2) # Greater than or equal to
```
