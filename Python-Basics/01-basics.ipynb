What is syntex?
Sytex refers to the set of rules that defines the combination of symbols that are considered to be correctly structured programs in a language.In simpler terms sytex is about the correct arrangement of words and symbols in a code.

What is Semantics?
Semantics refers to the meaning or the interpretation of the symbols,characters and commands in a language.It is about what the code is supposed to do when it runs.
