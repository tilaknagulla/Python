Importing Modules in Python:Modules and Packages

In python modules and packages help organize and reuse code.Hers's comprehensive guide on how to import them.

```python
import math
print(math.sqrt(4))
```

```python
from math import sqrt,pi
print(sqrt(4))
print(pi)
```

```python
import numpy as np
np.array([1,2,3,4,5,6])
```

```python
from math import *
print(sqrt(16))
print(pi)
```

```python
from packages.maths import addition
print(addition(1,2)) 
```

```python
from packages.subpackages.multiply import multiply
print(multiply(2,3))
```
