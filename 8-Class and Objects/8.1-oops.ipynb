#### Constructor:
A constructor in Python is a special method named __init__() that automatically runs when an object of a class is created.
Its main purpose is to initialize the object’s attributes and set up any required state for the object.

```python
### Instnace variables and methods
class Dog:
    # constructor
    def __init__(self,name,age):
        self.name=name
        self.age=age

#Create object
dog1=Dog("Tony",14)
print(dog1)
print(dog1.name,dog1.age)
dog2=Dog("Buddy",2)
print(dog2)
print(dog2.name,dog2.age)
```

```python
### Define classes and Instnace methods
class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def bark(self):
        print(f"The {self.name} says woof")

#Creating objects
dog1=Dog("Tony",14)
dog1.bark()
dog2=Dog("Lucy",3)
dog2.bark()
```

```python
### Modeling a Bank Account
## Define a class for bank account.

class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance


    def deposit(self,amount):
        self.balance+=amount
        print(f"{amount} is diposited,Now balance is {self.balance}")
    def withdraw(self,amount):
        if self.balance<amount:
            print(f"Insufficiant balance {self.balance}")
        else:
            self.balance-=amount
            print(f"{amount} withdrawn,Now balance is {self.balance}")

## Create the object
myaccount=BankAccount("Tilak")
myaccount.deposit(1000000000)
myaccount.withdraw(1)
```
