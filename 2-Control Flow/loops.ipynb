```python
# for loop
for i in range(5):
    print(i)
```

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range(10,1,-1):
    print(i)
```

```python
#strings
name="Tilak"
for i in name:
    print(i)
```

```python
## while loop
i=0
while i<10:
    print(i)
    i=i+1
```

```python
#Loop control statements
#break statement    
#break statement is used to exit a loop prematurely when a certain condition is met. It allows you to break out of the loop and continue with the rest of the program.
i=1
for i in range(10,15):
    if (i==12):
        print("Breaking the loop at i =", i)
        break
    print(i)

```

```python
#continue 
#The continue statement is used to skip the current iteration of a loop and move on to the next iteration. It allows you to bypass certain code within the loop when a specific condition is met.
for i in range(1,10):
    if i==5:
        print("This is continue statement at i =", i)
        continue
    print(i)
```

```python
#pass
#The pass statement is a null statement that is used as a placeholder when a statement is required syntactically but you don't want to execute any code. It allows you to create an empty block of code without causing a syntax error.
for i in range(1,10):
    if i==5:
        print("This is pass statement at i =", i)
        pass
    print(i)
```

```python
## nested loops a loops inside a loop
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
#Example calculate the sum of first n natural numbers
num=int(input("Enter the number:"))
sum=0
for i in range(num+1):
    if(i>0):
        sum=sum+i
print(f"The sum of first {num} natural numbers is {sum}")

#While loop
count=0
sum2=0
while count<=num:
    sum2=sum2+count
    count=count+1   
print(f"The sum of first {num} natural numbers is {sum2}")
```

```python
#Prime numbers between 1 to n
n=int(input("Enter the number:"))
for i in range(1,n+1):
    if i>1:
        for j in range(2,i):
            if i%j==0:
                break
        else:
            print(i)
```

```python
##Code for maths tables
num =int(input("Enter the number"))

for i in range(1, 11):
    if (i>0):
        print(i,"*",num,"=",i*num)
```
