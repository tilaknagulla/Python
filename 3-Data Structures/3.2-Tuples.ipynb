```python
#Immutable nature of tuple
#Tuples are immutable, meaning their elements can not be changed once assigned
lst=[1,2,3,4,5]
print(lst)
lst[1]="Tilak"
lst 
```

```python
numbers[1]="Tilak"
```

```python
#Tuple methods
print(numbers.count(1))
print(lst.count("Tilak"))
print(numbers.index(6))
```

```python
# Packing and unpacking tuple
packed_tuple=1,"Hello World",True
print(packed_tuple)
```

```python
#Unpacking a tuple
lst=[]
for pack in packed_tuple:
    lst.append(pack)

print(lst)
 

a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
#Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
#nested list
lst=[[1,2,3,4,],[5,6,7,8],[1,"Tilak",2.54,True]]
lst[1][0:3]
```

```python
#nusted tuple
nested_tuple=((1,2,3,4),('a','b','c','d'),(True,False))
#Access the elements inside the tuple
print(nested_tuple[1][3])
print(nested_tuple[2][1]) 
```

```python
#iterating over nested tuple
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```
