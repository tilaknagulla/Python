#### Inheritance:

Inheritance is a fundamental concept in OOPS that allows a class to inherit attributes and methods from another class. 

```python
### Single level inheritance

class Car:
    def __init__(self,windows,doors,engineType):
        self.windows=windows
        self.doors=doors
        self.enginetype=engineType
    
    def drive(self):
        print(f"The prson will drive the {self.enginetype} car")
    

```

```python
car1=Car(4,4,"petrol")
print(car1.drive())  
```

```python
class Car:
    def __init__(self,windows,doors,engineType):
        self.windows=windows
        self.doors=doors
        self.enginetype=engineType
    
    def drive(self):
        print(f"The prson will drive the {self.enginetype} car")

class Tesla(Car):
    def __init__(self,windows,doors,enginetype,is_selfdriving):
        super().__init__(windows,doors,enginetype)
        self.is_selfdriving=is_selfdriving
    def selfdriving(self):
        print(f"Tesla supports self driving : {self.is_selfdriving}")

tesla1=Tesla(6,6,"Electric",True)
tesla1.selfdriving()
print(tesla1.windows)
```

```python
## Multiple level inheritance
# Base class 1
class Animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print("Subclass must implement this method")
    def test(self):
        print("This is Animal Class")

# Base class 2

class Pet:
    def __init__(self,owner):
        self.owner=owner
    def test(self):
        print("This is Pet Class")

#Derived class
class Dog(Animal,Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
    def speak(self):
        return f"{self.name} says woof"
dog1=Dog("Tony","Syamala")
print(dog1.speak())
dog1.test()

        
```
