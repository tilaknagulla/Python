#### Wordnet Lemmatizer:

Lemmatization technique is like stemming.The output we will get after lemmatization is called lemma,Which is a root word ratherthan root stem,the output of stemming.After lemmatization,We will be getting a valid word that means the something.

NLTK provides WordNetLemmatizer class whcih is a thin wrapper around the wordnet corpus.This class uses morphy() function to the wordnet corpusreader class to find lemma.

```python
from nltk.stem import WordNetLemmatizer
lematizer = WordNetLemmatizer()
lematizer.lemmatize("Going",pos='a')
```

```python
from nltk.stem import WordNetLemmatizer
words=["eating","eat","eaten","going","goes","go","writing","writes","wrote","history","programming","programs","finally",
"finalize"]
lematizer=WordNetLemmatizer()
for word in words:
    print(word+"----->"+lematizer.lemmatize(word,pos='n'))
```
