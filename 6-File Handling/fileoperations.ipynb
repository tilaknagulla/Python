File Operation- Read and Write Files

File handling is s crucial part of any programming language.Python provides a build in functions method read from and write to files,both text and binary.This lesson will cover the basics of file handling,including reading and writing text files and binary files 

```python
with open('example.txt','r') as file:
    content=file.read()
    print(content)
    
```

```python
##Read a file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ## sstrip removes the newline character
```

```python
with open('example.txt','w') as file:
    file.write("Good morning\n")
    file.write('This is a new line')
    
```

```python
#Write a file without overwriting
with open('example.txt','a') as file:
    file.write('This is appending function\n')
    file.write('My name is : Tilak sai kumar Nagulla\n')
```

```python
#Writing a list of lines in to a file
lines=['\n','First line\n','Second line\n','Third line\n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
#Binary files 

#Wrting to a binary file
data=b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)

```

```python
#Reading binary file:
with open('example.bin','rb') as file:
    content=file.read()
    print(content)
```

```python
#Read the content from the source text file and write into a destination file
with open('example.txt','r') as file:
    read_content=file.read()
    with open('example_dst.txt','w') as file:
        file.write(read_content)
```

```python
##Read the txt file and count the number of lines words and characters
def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines =file.readlines()
        line_count=len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
        return line_count,word_count,char_count

file_path='example.txt'
lines,words,chars=count_text_file(file_path)
print(f'Line:{lines},Words:{words},Chars:{chars}')
```

```python
### Writing and then reading a file
with open('example.txt','w+') as file:
    file.write("Hello world\n")
    file.write("This is a new line \n")

    #move the file cursor to the begining
    file.seek(0)

    #read the content of the file
    content=file.read()
    print(content)
 
```

```python
#Is path exists
path="example.txt"
if os.path.exists(path):
    print(f"The path {path} exists")
else:
    print(f"The {path} does not exists")
```

```python
## Checking if the path is file or directory
import os

path="packages"
if os.path.isfile(path):
    print(f"The path {path} is file")
elif os.path.isdir(path):
    print(f"The path {path} is directory")
else:
    print(f"The path {path} is neither a file nor directory")
```

```python
## Getting the absolute path
relative_path = "example.txt"
absolute_path=os.path.abspath(relative_path)
print(absolute_path)
```
