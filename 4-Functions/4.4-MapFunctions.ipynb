The map() Function in Python:

The map() function applies a given function to all items in a list(or any other iterable) and returns a map object (an iterator).This is particularly useful for transforming data in a list comprehensively.

```python
def square(number):
    return number **2
```

```python
numbers=[2,3,4,5,6,9]
def square(number):
    return number **2
number_list=list((map(lambda x:x**2,numbers)))
#list(map(square,numbers))
print(number_list)
```

```python
#Map multiple iterables
numbers1=[1,2,3]
numbers2=[4,5,6]
added_numbers=list(map(lambda x,y:x+y,numbers1,numbers2))
print(added_numbers)
```

```python
#Map to convert string to list
str_numnbers=['1','2','3','4']
int_numbers=list(map(int,str_numnbers))
print(int_numbers) #[1,2,3,4]
```

```python
words=['apple','banana','cherry']
upper_word=list(map(str.upper,words))
print(upper_word)
```

```python
def get_name(person):
    return person['name']

people=[{'name':"Tilak","age":27},
        {'name':"Syamala","age":29}]

list(map(get_name,people))
```
