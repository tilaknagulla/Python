Real-World examples using lists in python:

Lists are one of the most commonly used data structures in python, thanks to their varstality and ease of use.

Here are several practical examples that illustrate their use in real-world scenarios

Example 1: Manage A To Do List

Create a To Do list to keep track of tasks

```python
to_do_list=["Take Bath","Have Breakfast","Watch Movie","Sleep for 2hr"]
##Add to task
to_do_list.append("Have amla")
to_do_list.append("Have dinner")

#Removing a completed tasks
to_do_list.remove("Have Breakfast")

#Check is there in the list or not
if "Have Breakfast" in to_do_list:
    print("Don't forgot to take the breakfast")

print("Todo list remaining")
for task in to_do_list:
    print(f"-{task}")
```

Example 2:Organizing Student Grades
Craete a list to store and calculate the average grades of the sudent

```python
#Organizing student grades
grades=[85,90,75,60,50]

#Adding grades
grades.append(95)
grades.append(95)
average_grade=sum(grades)/len(grades)
print(f"Average Grade:{average_grade:.2f}")
print("Highest grade:",max(grades))
print("Lowest grade:",min(grades))
```
