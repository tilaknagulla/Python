Data Types

1.Definition

  Datatypes are a classification of data which tell the complier or interpreter how the programmer intends to use the data.

  They determine the type of opertions that cen be performed on the data,the values that the data can take,and the amount of memory needed to store the data.
2.Importance of Data Types in Programming
   
   1.Data Types ensures the data is stored in an efficient way.
   2.They help in performing correct opertaions on data
   3.Proper use of data type can prevent errors and bugs in in a program
