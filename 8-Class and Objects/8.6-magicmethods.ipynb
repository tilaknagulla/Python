#### Magic Methods:

Magic methods in python is also known as bunder methods(Double underscore methods),are special methods that start and end with double underscores.These methods enable you to define the behaviour of objects for built in operations,such as arithmetic operations,comparisons and more.

Magic methods are predefined methods in python that you can override to change the behavior of your objects.Somemagic methods include:

- __init__:Initialize a new instnace of a class
- __str__ : Returns a string representation of an object
- __repr__ : Returns an official string representation of an object.
- __len__ : Returns an length of an object.
- __getitem__ : Gets an item from a container.
- __setitem__: Sets an item in a container.

```python
# Basic methods
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def __str__(self):
        return f"{self.name}.{self.age} years old"
    def __repr__(self):
        return f"Person name ={self.name} and age={self.age}"

person=Person("Krish",24)
print(person)
print(repr(person))
```
