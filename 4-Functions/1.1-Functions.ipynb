```python
def even_or_odd(num):
    '''This function finds even or odd'''
    if(num%2==0):
        print(f"- {num} is a even  number")
    else:
        print(f"- {num} is a odd number")


even_or_odd(88)
```

```python
#function with multiple parameters
def add(a,b):
    return a+b

result=add(5,4)
print(result)
```

```python
#Default parameters

def wish(name="Guest"):
    print("Hello", name)

wish("Tilak")
```

```python
#Variable length arguments
#Positional and keyword arguments
def print_numbers(*args):
    for arg in args:
        print(f"-{arg}")

print_numbers(1,2,3,4,4,5,"Tilak",4.14,True)

```

```python
#Keyword arguments
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"-{key} :{value}")

print_details(name="Tilak",age=27,address="Elamarru")
```

```python
#positional args and keyword args

def print_details(*args, **kwargs):
    for arg in args:
        print(f"-{arg}")
    for key, value in kwargs.items():
        # use a proper f‑string, include the "-" inside the quotes
        print(f"-{key}:{value}")

print_details(1, 2, 34, 5, 5, 6, name="Tilak", Address="Elamarru")
```

```python
#Return statements
def multiply(a,b):
    return a*b

print(multiply(2,3))
```
