Variables
========
Variables are the fundamental elements in programming used to store data that can be referenced and manipulated in a program.In python variables are created when you assign a value to them,and they do not need explicit declartion to reserve memory space.The declartion happens automatically when you assign a value to a variable.

```python
##Declaring and assigning variables
age =27
height=6.11
name="Tilak"
is_Student=False

##Printing the variables
print("Agev ",age)
print("Height ",height)
print("Name ",name)     
print("Is Student ",is_Student)
```

```python
##Naming Conventions
#1. Variable names should be descriptive and meaningful.    
#2. Variable names should start with a letter (a-z, A-Z) or an underscore (_).
#3. Variable names can contain letters, digits (0-9), and underscores (_), but cannot start with a digit.
#4. Variable names should not be a reserved keyword in Python (e.g., if, else, for, while, etc.).
#5. Variable names should be case-sensitive (e.g., age and Age are different variables).
#6. Variable names should not contain spaces. Use underscores (_) to separate words if needed (e.g., first_name).
#7. Variable names should be concise but descriptive enough to convey their purpose (e.g., total_price instead of tp).
#8. Variable names should follow a consistent naming convention (e.g., snake_case, camelCase, etc.) throughout the codebase.
## Valid variable names
first_name="Tilak"
lastName="Kumar"
```

```python
#Invalid variable names
#2age=30
#first-name="Tilak"
#@name="Tilak"
```

```python
#Understanding Variable Types
# Python is dynamically typed, which means you don't need to declare the type of a variable when you create it. The type is inferred based on the value assigned to the variable. Here are some common variable types in Python:
age=27 #integer
height=5.11 #float
name="Tilak" #string
is_Student=False #boolean

print(type(name)) #Output: <class 'str'>
```
