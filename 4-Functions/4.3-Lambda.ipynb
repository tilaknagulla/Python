Lambda Function:


Lambda functions are small anonymous function defined using the lambda keyword.They can have any number of arguments but only one expression.They are commonly used for short operations or as arguments to higher-order functions

```python
#syntex 
lambda arguments:expression
```

```python
addition=lambda a,b:a+b
print(addition(2,3))
```

```python
#Check even number with lambda function
even_num=lambda num:num%2==0

print(even_num(6))
```

```python
# map()
number=[2,3,47,8]
def square(number):
    return number**2

print(square(2))

```

```python
list(map(lambda x:x **2,number))
```
