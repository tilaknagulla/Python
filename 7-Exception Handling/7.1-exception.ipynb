Understanding Exceptions:

Exception handling in python allowes you to handle the errors gracefully and take corrective actions without stopping the execution of the program

What are Exception?

Exceptions are events that disrupt the normal flow of a program.They occur when an error is encountered during program execution.Common exceptions include.

1.ZeroDivisonError:Dividing by Zero
2.FileNotFoundError : File Not Found
3.ValueError : Invalid value
4.TypeError : Invalid Type

```python
## Exception try and except block
try:
    a=b
except:
    print("Variable has not been assigned")
```

```python
try:
    a=b
except NameError as ex:
    print(ex)
```

```python
try:
    result=1/2
    a=b
except ZeroDivisionError as ex1:
    print(ex1)
    print("Please enter the denominator greater than zero")
except Exception as ex2:
    print(ex2)
    print("Name b is not defined")
```

```python
## try,except,else and finally block
try:
    num=int(input("Please enter the name"))
    result=10/num

except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("The dineominator should be greaterthan 0")
except Exception as ex:
    print(ex)
else:
    print(f"The result is {result}")
finally:
    print("Execution is completed")
```
