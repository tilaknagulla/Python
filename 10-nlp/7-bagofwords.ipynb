#### Bag of Words — Definition
Bag of Words (BoW) is a text‑representation technique that converts text into numerical features by counting how many times each word appears in a document.

It ignores:

grammar

word order

sentence structure

It only cares about word frequency.

BoW is mainly used for:

text classification (spam detection, sentiment analysis)

document similarity

feature extraction for ML models



```python
Simple Example (Very Easy)
Input sentences:
Code
Sentence 1: "Cat sat on the mat"
Sentence 2: "Dog sat on the log"
Vocabulary:
Code
["cat", "sat", "on", "the", "mat", "dog", "log"]
Bag of Words vectors:
Word →	cat	sat	on	the	mat	dog	log
S1	1	1	1	1	1	0	0
S2	0	1	1	1	0
```

```python
import pandas as pd
messages=pd.read_csv('dataset.csv',names=['label','message'])
print(messages)
```

```python
## Data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer 
from nltk.stem import WordNetLemmatizer
wn=WordNetLemmatizer()
ps=PorterStemmer()
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['label'][i])
    review=review.lower()
    review=review.split()
    #review=[ps.stem(word) for word in review if word not in stopwords.words('english')]
    #applying lemmatization
    review=[wn.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
print(corpus)
```

```python

## create a bag of words model
from sklearn.feature_extraction.text import CountVectorizer
cv =CountVectorizer(max_features=7,binary=True)
x=cv.fit_transform(corpus).toarray()
x
```

#### Definition of N‑grams
An N‑gram is a sequence of N consecutive words (or characters) taken from a text.

If N = 1 → Unigram

If N = 2 → Bigram

If N = 3 → Trigram

If N = 4 → 4‑gram, and so on

N‑grams help capture context, word order, and local patterns in text, unlike Bag of Words which ignores order.

Example :
Sentance="I love natural language processing"
unigram=["I", "love", "natural", "language", "processing"]
bigram=["I love", "love natural", "natural language", "language processing"]
trigram=["I love natural", "love natural language", "natural language processing"]


Why N‑grams are useful
They help capture:

Context (e.g., “New York” is different from “New” + “York”)

Phrase patterns (e.g., “not good” → negative sentiment)

Word order, which Bag of Words loses

Used in:

Text classification

Language modeling

Spam detection
Autocomplete

Machine translation

```python
#create bag of words with ngram
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=7,binary=True,ngram_range=(3,3))
X=cv.fit_transform(corpus).toarray()
cv.vocabulary_
```
