Dictionaries:

Dictionaries are unordered collection of items.They store data in key-value pairs.Keys must be uniqu and immutable(eg,.String,Numbers,or Tuples),while values can be of any type.

```python
#Define Dictionaries
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
print(empty_dict)
```

```python
student={"name":"Tilak","age":27,"grade":'A'}
print(student)
```

```python
#Always use single key
student={"name":"Tilak","age":27,"grade":'A',"age":30}
print(student)
```

```python
#Accessing dictionary elements
student={"name":"Tilak","age":27,"grade":'A'}
print(student)
```

```python
print(student["grade"])
print(student["age"])

#Accessing through get() method
print(student.get('grade'))
print(student.get('last_name'))

#Assign some default values
print(student.get('last_name',"Not Available"))
```

```python
#Modifying Dictionary Elements
## Dictionary are mutable so you can add,update or delete elements
print(student)
student["grade"]='A++' # Update the value for the key
student["address"]="India" #Add the new key and value
print(student)
```

```python
#Dictionry methods
keys=student.keys() ##get all the keys
print(keys)
values=student.values() ##get all the values
print(values)
items=student.items()  ## get all the items
print(items)
```

```python
## shallow copy
student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="Sai"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy() ##shallow copy
print(student_copy1)
print(student)
student["name"]="Tilak"
print(student)
```

```python
#Iterating over Dictionaries
# You can use loops to iterate over dictionaries,keys,values ,or items

#Iterating over keys
for keys in student.keys():
    print(keys)
print("=============================================")
#Iterating over values
for values in student.values():
    print(values)
print("=============================================")
#for iterating over items
for key,value in student.items():
    print(f"{key}:{value}")

```

```python
#Nested dictinaries
students={
      "student1":{"name":"Krish","age":27},
      "student2":{"name":"Kumar","age":28}
}
print(students)
```

```python
#Access nested dictionaries elements
print(students["student1"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
## Iterating over nested dictionaries

for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key, value in student_info.items():
        print(f"{key}:{value}")
```

```python
##Dictionaries Comprehension
squares={x:x**2 for x in range(10)}
print(squares)
```

```python
#Condition dictionary comprehension
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
#Practical examples
## Use a dictinary to count the frequency of elements in a list

numbers=[1,2,2,3,3,4,9,1,3,4,5,6,4,32]
frequency={}
for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
       frequency[number]=1
print(frequency)
```

```python
##Merge two dictionaries into one
dict1={"a":1,"b":2}
dict2={"b":3,"c":4}

merge_dict={**dict1,**dict2}
print(merge_dict)
```
