#### Stemming:
Stemming is a process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as lemma.Stemming is important in natural language understanding(NLU) and Natural language processing(NLP).

```python
## Classification problem
## Comments of a product is a poitive review or negative review
## Review --> eating,eat,eaten [going,gone,goes]-->go
words=["eating","eat","eaten","going","goes","go","writing","writes","wrote","history","programming","programs","finally",
"finalize"]
```

```python
## Porterstemmer

from nltk.stem import PorterStemmer
words=["eating","eat","eaten","going","goes","go","writing","writes","wrote","history","programming","programs","finally",
"finalize"]
stemming=PorterStemmer()
for word in words:
    print(word+"---->"+stemming.stem(word))
```

#### Regexpstemmer class :
NLTK has regexpstemmer class with the help of which we acn easily implement regular expression stemmer algorithms.It basically takes a single regular expressiin and removes and prefix or suffix that matches the expression.

```python
from nltk.stem import RegexpStemmer
regex_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
regex_stemmer.stem("eating")
```

```python
print(regex_stemmer.stem('ingeating'))
```

### Snowball stemmer:
The Snowball Stemmer is an improved, more advanced version of the classic Porter Stemmer used in Natural Language Processing (NLP) to reduce words to their root or stem form.

Think of it as a smart word‑shortening tool that helps algorithms treat similar words as the same.

```python
from nltk.stem import SnowballStemmer
words=["eating","eat","eaten","going","goes","go","writing","writes","wrote","history","programming","programs","finally",
"finalize"]
snowball_stemmer=SnowballStemmer("english")
for word in words:
    print(word+"---->"+snowball_stemmer.stem(word))
```
